# Load SM DBs

```r
# Load and prepare Gaulton DB
Gaulton.db.dir = "/nfs/lab/Luca/Assets/LR.Database/GaultonDB_V3.txt"
# Load list of ligands and receptors
SM.ls.df = read.table(paste(assets.dir, "SM.ls.df.txt", sep=""), sep = "\t", stringsAsFactors = F, header = T)
SM.ls = unique(SM.ls.df$gene)
length(SM.ls)

# Load database
cc.db.data = read.table(cc.db, sep = "\t", stringsAsFactors = F, header = T)
# cc.db.data = cc.db.data[,c(11,2,3)]
message("CCDB number of interactions: ", length(levels(factor(cc.db.data$interaction_name))))

gaulton.db.data = read.table(Gaulton.db.dir, sep = "\t", stringsAsFactors = F, header = T)
colnames(gaulton.db.data)[1] = "gene"

SM.ls.df.GDB = merge(SM.ls.df, gaulton.db.data, by = "gene", all.x = TRUE)
SM.ls.df.GDB[is.na(SM.ls.df.GDB)] <- ""

# Calculate category sizes
categories.size = SM.ls.df.GDB[SM.ls.df.GDB$Type == "Ligand",]

# Ligands
ligands.GDB = filter(SM.ls.df.GDB, Type != "Receptor")
colnames(ligands.GDB)[1] = "ligand"
head(ligands.GDB)

# Make a list of all ligands
ligands.ls = c(unique(cc.db.data$ligand), 
              str_split_fixed(unique(cc.db.data$ligand[grepl("_", cc.db.data$ligand)]), "_", n = 2)[,1],
              str_split_fixed(unique(cc.db.data$ligand[grepl("_", cc.db.data$ligand)]), "_", n = 2)[,2])
ligands.ls = unique(ligands.ls)
ligands.ls = ligands.ls[ligands.ls %in% SM.ls]
# Make a list of all receptors
receptors.ls = c(unique(cc.db.data$receptor),
                 str_split_fixed(unique(cc.db.data$receptor[grepl("_", cc.db.data$receptor)]), "_", n = 2)[,1],
                 str_split_fixed(unique(cc.db.data$receptor[grepl("_", cc.db.data$receptor)]), "_", n = 2)[,2])
receptors.ls = unique(receptors.ls)
receptors.ls = receptors.ls[receptors.ls %in% SM.ls]

# Make a list of all receptors that are also ligands
both = ligands.ls[ligands.ls %in% receptors.ls]
message("Number of ligands that are also receptors: ", length(both))


# Exclude those from the other lists
receptors.ls = receptors.ls[!receptors.ls %in% both]
message("Number of receptors: ", length(receptors.ls))


# Exclude those from the other lists
ligands.ls = ligands.ls[!ligands.ls %in% both]
message("Number of ligands: ", length(ligands.ls))


#Check that the length is 860
message("Final Number of SM: ", length(ligands.ls) + length(receptors.ls) + length(both))

Interaction.type.ls = c("Secreted Signaling", "Cell-Cell Contact", "ECM-Receptor")

Functional.classification.order = c('Hormones & Neuropeptides', 'Enzymes', 'Developmental Proteins', 
                                    'Growth Factors and Cytokines', 'Antigen-Presenting Molecules', 'Complement System',
                                    'Immune regulators', 'Modulators', 'Cell Adhesion Molecule', 'ECM Proteins')
```

# Build consensus of interactions

```r
setwd(results.dir)
LR.list <- dir(pattern = "wholeccDB.txt")
i <- LR.list[1]
Prefix <- strsplit(i, "_df.net")[[1]][1]
message("Processing: ", Prefix)
LR.data <- read.table(i, sep = "\t", stringsAsFactors = F, header = T)
LR.data$condition = Prefix
LR.data.consensus <- LR.data
message("Detected: ", nrow(LR.data.consensus), " interactions")

LR.list <- LR.list[-1]
for (i in LR.list){
    Prefix <- strsplit(i, "_df.net")[[1]][1]
    message("Processing ", Prefix)
    LR.data <- read.table(i, sep = "\t", stringsAsFactors = F, header = T)
    LR.data$condition = Prefix
    message("Detected: ", nrow(LR.data), " interactions")
    LR.data.consensus <- rbind(LR.data.consensus, LR.data)
}

write.table(LR.data.consensus, paste(results.dir, "LR.data.consensus.txt", sep = ""), 
            sep = "\t", row.names = F, col.names = T)
```

# Subset

## FDR Correction

```r
sig.tresh = 0.1
```

```r
# FDR correction
LR.data.consensus$pval[LR.data.consensus$pval == 0] = 0.0001 # replacing 0s with a low number
options(repr.plot.width=10, repr.plot.height=5)
hist(LR.data.consensus$pval, main = "Pvalue distribution")
message("all LR pairs: ", nrow(LR.data.consensus))
message("ns LR pairs (pvalue > 0.05): ", length(LR.data.consensus$pval[LR.data.consensus$pval > 0.05]))
message("LR pairs (pvalue < 0.05): ", length(LR.data.consensus$pval[LR.data.consensus$pval < 0.05]))
message("LR pairs (pvalue < 0.01): ", length(LR.data.consensus$pval[LR.data.consensus$pval < 0.01]))
message("LR pairs (pvalue = 0): ", length(LR.data.consensus$pval[LR.data.consensus$pval == 0.001]))
message("LR pairs (pvalue < 0.001): ", length(LR.data.consensus$pval[LR.data.consensus$pval < 0.001]))
LR.data.consensus$FDR = p.adjust(LR.data.consensus$pval, 'BH')
hist(LR.data.consensus$FDR,  main = "FDR distribution")
message("all LR pairs: ", nrow(LR.data.consensus))
message("ns LR pairs (FDR > 0.1): ", length(LR.data.consensus$FDR[LR.data.consensus$FDR > 0.1]))
message("LR pairs (FDR < 0.1): ", length(LR.data.consensus$FDR[LR.data.consensus$FDR < 0.1]))
message("LR pairs (FDR < 0.05): ", length(LR.data.consensus$FDR[LR.data.consensus$FDR < 0.05]))
message("LR pairs (FDR < 0.01): ", length(LR.data.consensus$FDR[LR.data.consensus$FDR < 0.01]))
message("LR pairs (FDR < 0.001): ", length(LR.data.consensus$FDR[LR.data.consensus$FDR < 0.001]))
```

```r
options(repr.plot.width=10, repr.plot.height=5)
ggplot(LR.data.consensus, aes(x = FDR))+
    geom_histogram(binwidth=0.01, color = "black", fill = "#0099f9", alpha = 0.75) +
    stat_bin(geom="text", breaks = c(0, 0.00999, 1.01), position=position_stack(vjust=0.5), aes(label=ifelse(..count..>1, ..count.., ""))) +
    labs(y= "Count", x = "FDR", title = "All CellChat interactions") +
    theme_bw() +
    ylim(0, 100000) +
    facet_zoom(xlim = c(-0.01, 0.01)) +
    theme(axis.text = element_text(size = 12), axis.title = element_text(size = 12, face = "bold"), 
                         plot.title = element_text(size = 18, face = "bold", , hjust = 0.5)) +
    geom_vline(aes(xintercept=0.05),
            color="black", linetype="dashed", size=0.5)
```

```r
# Subset significant interactions
LR.data.consensus.sig = LR.data.consensus[LR.data.consensus$FDR < sig.tresh, ]
message("Sig LR pairs: ", nrow(LR.data.consensus.sig))
```

```r
options(repr.plot.width=10, repr.plot.height=5)
ggplot(LR.data.consensus.sig, aes(x = pval))+
    geom_histogram(binwidth=0.01, color = "black", fill = "#0099f9", alpha = 0.75) +
    stat_bin(geom="text", breaks = c(0, 0.00999, 1.01), position=position_stack(vjust=0.5), aes(label=ifelse(..count..>1, ..count.., "")))
```

## Set interaction strenght threshold

```r
IS.tresh = quantiles[2]
#IS.tresh = 0.00005
#IS.tresh = 0.005
```

## Decide the IS subset

```r
# ALL DATA
# LR.data.consensus.sig.thresh = LR.data.consensus

# JUST PADJ filtering
# LR.data.consensus.sig.thresh = LR.data.consensus.sig

# Subset for IS 
LR.data.consensus.sig.thresh = LR.data.consensus.sig[LR.data.consensus.sig$prob > IS.tresh, ]
```

```r
message("Final # of LR pairs: ", nrow(LR.data.consensus.sig.thresh), 
        "\n FDR theshold used: ", sig.tresh,
        "\n IS theshold used: ", IS.tresh)
```

```r
# ADD interaction ID columns
LR.data.consensus.sig.thresh$interaction.id = paste(LR.data.consensus.sig.thresh$source, 
                                  LR.data.consensus.sig.thresh$target, 
                                  LR.data.consensus.sig.thresh$ligand,
                                  LR.data.consensus.sig.thresh$receptor,
                                  sep = "--")
```

```r
cell.pop.order
```

```r
# Blacklist Ligands:
Blacklist.genes = c("INS")
Whitelist.pops = c("Beta")
Blacklist.pops = cell.pop.order[!cell.pop.order %in% Whitelist.pops]
message("Removing ", Blacklist.genes, " from:")
print(Blacklist.pops)
LR.data.consensus.sig.thresh = LR.data.consensus.sig.thresh[!c(LR.data.consensus.sig.thresh$source %in% Blacklist.pops & LR.data.consensus.sig.thresh$ligand %in% Blacklist.genes),]

Blacklist.genes = c("GCG")
Whitelist.pops = c("Alpha")
Blacklist.pops = cell.pop.order[!cell.pop.order %in% Whitelist.pops]
message("Removing ", Blacklist.genes, " from:")
print(Blacklist.pops)
LR.data.consensus.sig.thresh = LR.data.consensus.sig.thresh[!c(LR.data.consensus.sig.thresh$source %in% Blacklist.pops & LR.data.consensus.sig.thresh$ligand %in% Blacklist.genes),]

Blacklist.genes = c("SST")
Whitelist.pops = c("Delta")
Blacklist.pops = cell.pop.order[!cell.pop.order %in% Whitelist.pops]
message("Removing ", Blacklist.genes, " from:")
print(Blacklist.pops)
LR.data.consensus.sig.thresh = LR.data.consensus.sig.thresh[!c(LR.data.consensus.sig.thresh$source %in% Blacklist.pops & LR.data.consensus.sig.thresh$ligand %in% Blacklist.genes),]

Blacklist.genes = c("PRSS1", "PRSS2", "PRSS3")
Whitelist.pops = c("Acinar1_2_6", "Acinar_3", "Acinar_4", "Acinar_5")
Blacklist.pops = cell.pop.order[!cell.pop.order %in% Whitelist.pops]
message("Removing ", Blacklist.genes, " from:")
print(Blacklist.pops)
LR.data.consensus.sig.thresh = LR.data.consensus.sig.thresh[!c(LR.data.consensus.sig.thresh$source %in% Blacklist.pops & LR.data.consensus.sig.thresh$ligand %in% Blacklist.genes),]

# Receptors
Blacklist.genes = c("CD8A", "CD8B", "CD8B2")
Whitelist.pops = c("Tcells")
Blacklist.pops = cell.pop.order[!cell.pop.order %in% Whitelist.pops]
message("Removing ", Blacklist.genes, " from:")
print(Blacklist.pops)
LR.data.consensus.sig.thresh = LR.data.consensus.sig.thresh[!c(LR.data.consensus.sig.thresh$target %in% Blacklist.pops & LR.data.consensus.sig.thresh$receptor %in% Blacklist.genes),]

Blacklist.genes = c("CD4")
Whitelist.pops = c("Tcells", "Macrophage")
Blacklist.pops = cell.pop.order[!cell.pop.order %in% Whitelist.pops]
message("Removing ", Blacklist.genes, " from:")
print(Blacklist.pops)
LR.data.consensus.sig.thresh = LR.data.consensus.sig.thresh[!c(LR.data.consensus.sig.thresh$target %in% Blacklist.pops & LR.data.consensus.sig.thresh$receptor %in% Blacklist.genes),]


message("# of LR pairs: after blacklist: ", nrow(LR.data.consensus.sig.thresh))
```

```r
levels(factor(LR.data.consensus.sig.thresh$source))
levels(factor(LR.data.consensus.sig.thresh$target))
```

```r
# Set order for sources
LR.data.consensus.sig.thresh$source = factor(LR.data.consensus.sig.thresh$source, levels = c(cell.pop.order))

# Assign compartments
LR.data.consensus.sig.thresh$compartment.source = LR.data.consensus.sig.thresh$source

LR.data.consensus.sig.thresh$compartment.source = ifelse(LR.data.consensus.sig.thresh$compartment.source %in%  Endocrine, "Endocrine", 
                 ifelse(LR.data.consensus.sig.thresh$compartment.source %in%  Exocrine, "Exocrine",
                        ifelse(LR.data.consensus.sig.thresh$compartment.source %in%  Immune, "Immune",
                               ifelse(LR.data.consensus.sig.thresh$compartment.source %in%  Endothelial, "Endothelial", "Stromal"))))

LR.data.consensus.sig.thresh$compartment.source = factor(LR.data.consensus.sig.thresh$compartment.source, levels = c('Endocrine','Exocrine','Immune','Endothelial', "Stromal"))

# Set order for targets
LR.data.consensus.sig.thresh$target = factor(LR.data.consensus.sig.thresh$target, levels = c(cell.pop.order))

# Assign compartments
LR.data.consensus.sig.thresh$compartment.target = LR.data.consensus.sig.thresh$target

LR.data.consensus.sig.thresh$compartment.target = ifelse(LR.data.consensus.sig.thresh$compartment.target %in%  Endocrine, "Endocrine", 
                 ifelse(LR.data.consensus.sig.thresh$compartment.target %in%  Exocrine, "Exocrine",
                        ifelse(LR.data.consensus.sig.thresh$compartment.target %in%  Immune, "Immune",
                               ifelse(LR.data.consensus.sig.thresh$compartment.target %in%  Endothelial, "Endothelial", "Stromal"))))

LR.data.consensus.sig.thresh$compartment.target = factor(LR.data.consensus.sig.thresh$compartment.target, levels = c('Endocrine','Exocrine','Immune','Endothelial', "Stromal"))

LR.data.consensus.sig.thresh = merge(LR.data.consensus.sig.thresh, ligands.GDB, by = "ligand")
```
